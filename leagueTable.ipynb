```python
import requests as re
from bs4 import BeautifulSoup
```

```python
url = "https://fbref.com/en/comps/9/Premier-League-Stats#all_results2023-202491"

page = re.get(url)

soup_obj = BeautifulSoup(page.text,'html.parser')

#print(soup_obj)
```

```python
table = soup_obj.find('table')

#print(table)
```

```python
headers = soup_obj.find_all('th')
#print(headers)
```

```python
table_headers = [header.text.strip() for header in headers]

#print(table_headers)
```

```python
# Extracting column headers
final_headers = []
for header in table_headers:
    if "\n" in header:
        final_headers.append(header.split("\n")[0])
        del final_headers[19:]
    else:
        final_headers.append(header)
        del final_headers[19:]

#print(final_headers)
```

```python
import pandas as pd

# Assigning headers from table as dataframe column names
df = pd.DataFrame(columns=final_headers)
df
```

```python
column_data = table.find_all('tr')


for  row in column_data[1:]:
    # extracting team position
    position = row.find('th').text.strip() if row.find('th') else None
    
    # extracting team data
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    del individual_row_data[18:]

    # add extracted data to the dataframe 
    length = len(df)
    df.loc[length] = [position] + individual_row_data

df
      
```
